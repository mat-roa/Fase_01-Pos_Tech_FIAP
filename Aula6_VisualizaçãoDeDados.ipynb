```python
#Criando um dicionario meses{} para poder converter o mês do index "2023/Mai" para um numero correspondente
# : str e : int para indicar qual sera o tipo da variavel

from datetime import date

meses = {
    "Jan" : 1,
    "Fev" : 2,
    "Mar" : 3,
    "Abr" : 4,
    "Mai" : 5,
    "Jun" : 6,
    "Jul" : 7,
    "Ago" : 8,
    "Set" : 9,
    "Out" : 10,
    "Nov" : 11,
    "Dez" : 12
}

def para_dia(ano_mes: str):
  ano: int = int(ano_mes[:4])
  mes_como_string = ano_mes[5:]
  mes: int = meses[mes_como_string]
  return date(ano, mes, 1)
```

```python
mensal.index = mensal.index.map(para_dia)
```

```python
#Escolhendo 3 estados para plotar
estados = ["Amazonas","Mato Grosso","Ceará"]
por_ano_estados = mensal_aberto.query("uf in @estados").groupby(['uf','ano']).sum().reset_index()
por_ano_estados
```

```python
sns.barplot(data=por_ano_estados, x = 'ano', y ='gasto', hue='uf')
```

```python
#catplot() plota varios graficos na mesma figura
#col compartilha o eixo Y, row compartilha o eixo X
sns.catplot(data=por_ano_estados, x = 'ano', y ='gasto', col='uf', kind='bar')

plt.grid()
```
